# NYT Best Sellers Dataframe 

```python
import requests
import json
import pandas as pd
```

### Get genre names list to use for master dataframe

```python
r_genre = requests.get("https://api.nytimes.com/svc/books/v3/lists/names.json?api-key=2db89f3514744a55905084f3441989ca")
r_genre.json()
```

```python
genre_list = r_genre.json()['results']
len(genre_list)
```

```python
genres = []
for i in range(len(genre_list)):
    genres.append(genre_list[i]['list_name_encoded'])

# this is a list of all the genres of the NYT best sellers
genres
```

### Methods to get all NYT best selling books for book dataframe

```python
# we will split genres in order to call the API, by break the genre list into parts
genres_1 = genres[:11]
genres_2 = genres[11:22]
genres_3 = genres[22:33]
genres_4 = genres[33:44]
genres_5 = genres[44:55]
genres_1

# we will run each subset of genres individually to make the dataframe
```

```python
# method to get request all the genres
def make_urls_by_genre():
    genre_urls = []
    for i in genres_1:
        link = 'https://api.nytimes.com/svc/books/v3/lists/'+i+'/2019-01-06.json?api-key=2db89f3514744a55905084f3441989ca'
        genre_urls.append(link)
    return genre_urls
```

```python
make_urls_by_genre()
```

```python
# will break up into 
def api_calls_by_genre():
    mini_list = []
    for i in make_urls_by_genre():
        r = requests.get(i)
        mini_list.append(r.json())
    return mini_list
```

```python
all_books_list = api_calls_by_genre()
```

```python
bookbook = []
for j in range(int(len(all_books_list))):
    for i in range(len(all_books_list[j]['results']['books'])):
        bookbook.append({'Title':all_books_list[j]['results']['books'][i]['title'], 'Author':all_books_list[j]['results']['books'][i]['author'], 'Publisher':all_books_list[j]['results']['books'][i]['publisher'], 'Description':all_books_list[j]['results']['books'][i]['description'], 'Genre':all_books_list[j]['results']['list_name_encoded']})
    
# we had for j in range(int(len(all_books_list)/2)):
#which works, but only returns half the list
```
